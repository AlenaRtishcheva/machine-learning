**Операции с производными**


```python
import numpy as np
import sympy as sp
```

***Нахождение производной функции***

```python
# Создаем символьную переменную
x = sp.Symbol("x")

# Определяем функцию
f = sp.cos(x**2) * 5 * x ** 3

# Находим производную функции
df = sp.diff(f, x)

print(df)
```

***Нахождение производной функции в точке***

```python
# Опеределим функцию
def f(x):
    return x**2

# Определим точку в которой будем вычислять производную
x0 = 2

# Определим шаг дискретизации
h = 0.00001

# Вычислим конечную разность функции в точке x0
df_dx = (f(x0 + h) - f(x0)) / h

print(df_dx)
```

```python
# Создаем символьную переменную
x = sp.Symbol("x")

# Определяем функцию
f = x**2
x0 = 2

df_dx = f.diff(x)

# вычисляем производную функции в точке x0
df_dx_value = df_dx.subs(x, x0)

print(df_dx_value)
```
