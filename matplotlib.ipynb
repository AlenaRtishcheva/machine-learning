**Построение графиков**


```python
import numpy as np
import matplotlib.pyplot as plt
```

***f(x)=ax+b***

```python
x = np.linspace(-10, 10, 100)
a = 2
b = 1
y = a * x + b
plt.plot(x, y, color="blue", linestyle="-", linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции y = ax + b")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True, alpha=0.3)  # Добавляем сетку

plt.show()
```

***f(x) = x^2***

```python
x = np.linspace(-10, 10, 100)
y = x ** 2
plt.plot(x, y, color="blue", linestyle="-", linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции y = x**2")
plt.xlim(-10, 10)
plt.ylim(-2, 10)
plt.grid(True, alpha=0.3)  # Добавляем сетку

plt.show()
```

**f(x)=1/x**

```python
x = np.linspace(-10, 10, 200)
x1 = x[:100]
x2 = x[101:]

# Вычисление значений y, за исключением точки x=0
y1 = 1 / x1
y2 = 1 / x2

plt.plot(x1, y1, color="blue", linestyle="-", linewidth=2)
plt.plot(x2, y2, color="blue", linestyle="-", linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y")
plt.title("График функции y = 1/x")
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(True, alpha=0.3)  # Добавляем сетку

plt.show()
```
